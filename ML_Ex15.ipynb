# Matplotlib Figure Object

```python
import numpy as np
```

```python
a = np.linspace(0,10,11)
```

```python
a
```

```python
b = a**4
b
```

```python
x = np.arange(0,10)
y = 2 * x
```

```python
x
```

```python
y
```

```python
import matplotlib.pyplot as plt
```

```python
fig = plt.figure()
```

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1]) ## left, bottom, widht, height

axes.plot(x,y)

plt.show()
```

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1]) ## left, bottom, widht, height

axes.plot(a,b)

plt.show()
```

```python
type(fig)
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0,0,1,1])
axes2 = fig.add_axes([0.2,0.2,0.5,0.5])

axes1.plot(a,b)
axes1.set_xlabel('X-axis')
axes1.set_ylabel('Y-axis')
axes1.set_title('Big Plot')

axes2.plot(a,b)
axes2.set_title('Small Plot')
plt.show()
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0,0,1,1])
axes2 = fig.add_axes([0.2,0.5,0.25,0.25])

axes1.plot(a,b)
axes1.set_xlabel('X-axis')
axes1.set_ylabel('Y-axis')
axes1.set_title('Big Plot')

axes2.plot(a,b)
axes2.set_xlim(5,10)
axes2.set_ylim(4000,1000)
axes2.set_title('Zoom IN Plot')
fig.savefig('fig.png', bbox_inches='tight')
plt.show()
```
