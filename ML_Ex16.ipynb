# Matplotlib Subplots

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
a = np.linspace(0,10,11)
b = a**4
```

```python
x = np.arange(0,10)
y = 2 * x
```

## plt.subplots()

```python
fig, axes = plt.subplots()

axes.plot(x,y, 'g')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('Subplots')
```

```python
fig, axes = plt.subplots( nrows=1,ncols=3)
```

```python
axes
```

```python
axes.shape
```

```python
fig, axes = plt.subplots( nrows=2,ncols=2)
```

```python
axes
```

```python
axes.shape
```

```python
fig, axes = plt.subplots(nrows=1,ncols=2)

for axe in axes:
    axe.plot(x,y,'g')
```

```python
fig, axes = plt.subplots(nrows=1,ncols=2)

axes[0].plot(a,b,'m')
axes[1].plot(x,y,'r')
```

```python
fig, axes = plt.subplots(nrows=2,ncols=2)

axes[0][0].plot(a,b,'m')
axes[1][1].plot(x,y,'r')
axes[0][1].plot(a,b,'m')
axes[1][0].plot(x,y,'r')
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2)

axes[0][0].plot(a,b)
axes[1][1].plot(x,y)

plt.tight_layout()
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12,8))

axes[0][0].plot(a,b)
axes[0][0].set_title('Plot00')

axes[1][1].plot(x,y)
axes[1][1].set_title('Plot11')
axes[1][1].set_xlabel('xPlot11')
axes[0][1].plot(a,b,'m')
axes[1][0].plot(x,y,'r')

fig.suptitle("Multi Grid Subplot")
plt.show()
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12,8))

axes[0][0].plot(a,b,'m')
axes[1][1].plot(x,y,'r')
axes[0][1].plot(a,b,'m')
axes[1][0].plot(x,y,'r')

fig.savefig('subplots.png',bbox_inches='tight')

plt.show()
```
