## Conditional Filtering

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('tips.csv')
```

```python
df.head()
```

```python
df['total_bill'] > 30
```

```python
df[df['total_bill'] > 30]
```

```python
mask = df['total_bill'] > 30
df[mask]
```

```python
df[df['sex'] == 'Male']
```

```python
df[(df['total_bill'] > 30) & (df['sex'] == 'Male')]
```

```python
df[(df['total_bill'] > 30) & ~(df['sex'] == 'Male')]
```

```python
## Findout customers whos gender is Male and total bill is greater than 30$
df[(df['total_bill'] > 30) & (df['sex'] != 'Male')]
```

```python
## Findout customer visit either on Sun or Sat
df[(df['day'] == 'Sun') | (df['day'] == 'Sat')]
```
